# **withColumn() & withColumnRenamed()**

```python
display(df.limit(5))
```

```python
### Change the data type of column 
from pyspark.sql.functions import *
from pyspark.sql.types import *

df1 = df.withColumn('Quantity',col('Quantity').cast('Integer'))
display(df1.limit(3))
```

```python
### update the val of existing col 
df2=df1.withColumn('Quantity',col('Quantity')*2)
display(df2.limit(3))
```

```python
### create a new col from existing
display(df2.withColumn('Quantity_1',col('Quantity')/2))
```

```python
### add a new col 

df3= df2.withColumn('country1',lit('USA'))
display(df3.limit(2))
```

```python
### renamed he col name 
df4 = df3.withColumnRenamed('Ship_Mode','ship_mode')
display(df4.limit(2))
```

# **How to write the data and create delta tables**

```python
df = spark.read.format('csv').options(header='True',inferSchema=True).load('abfss://source_workspace@onelake.dfs.fabric.microsoft.com/source_lk.Lakehouse/Files/superstore.csv')
df=df.withColumnRenamed('Ship _Date','Ship_Date').withColumnRenamed('Product_ ID','Product_ID')
display(df.limit(4))
```

```python
-- lake house -- pyspark -- df --- file location , tables 
```

```python
df.write.format('delta').mode('append').option('path','abfss://datafactory@onelake.dfs.fabric.microsoft.com/destination.Lakehouse/Files/superstore/').save()
```

```python
-- append      --- location -- files
-- overwrite --- location --- file
-- error  --- location - error
-- ignore --- location -- 
```

```python
-- Managed table 
-- external table 
```

**MANAGED TABLE**

```python
df.write.format('delta').mode('append').saveAsTable('destination.superstore')
```

**EXTERNAL TABLE**

```python
df.write.format('delta').mode('append').option('path','abfss://datafactory@onelake.dfs.fabric.microsoft.com/destination.Lakehouse/Files/external').saveAsTable('external_table')
```
