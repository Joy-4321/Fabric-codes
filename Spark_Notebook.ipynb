**Changing the Data type, using the with column and with column rename()**

```python
df =spark.read.format('csv').options(header = 'True').load('abfss://sparks@onelake.dfs.fabric.microsoft.com/Spark_lake.Lakehouse/Files/sample_superstore.csv')
display(df)
```

```python
# changing the Data type from string to interger in one of the column. in this case the quantity column.
### import your libraries first.
from pyspark.sql.functions import *
from pyspark.sql.types import *

df1 =df.withColumn('Quantity',col('Quantity').cast('Integer'))

display(df1)

display(df1.limit(3))   # limiting the total number of rows i want to see.
```

```python
# updating the value of an existing column, here will multiply the value of the quantity column.

df2 =df1.withColumn('Quantity',col('Quantity')*2)

display(df2.limit(3))
```

```python
# Creating a new column in an already existing table column,and filing it with data.

df3 =df2.withColumn('Quantity_new',col('Quantity')/2)
display(df3.limit(3))
```

```python
# creating a a new column in a table and passing a value, we use lit to pass a value
df4 =df3.withColumn('New_country',lit('USA'))

display(df4)
```

```python
# renaming an existing column

df5 =df4.withColumnRenamed('Postal Code','Postal_Code')

display(df5)
```
